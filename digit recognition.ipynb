```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits


digits = load_digits()

X = digits.data
y = digits.target

print(f"Shape of features (X): {X.shape}")
print(f"Shape of targets (y): {y.shape}")
print(f"Number of classes: {len(set(y))}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

linear_model.fit(X_train, y_train)

print("Linear Regression model trained successfully.")
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

y_pred_continuous = linear_model.predict(X_test)

y_pred_discrete = np.round(y_pred_continuous)
y_pred_discrete = np.clip(y_pred_discrete, 0, 9).astype(int)

accuracy = accuracy_score(y_test, y_pred_discrete)

print(f"Model Accuracy: {accuracy:.4f}")
```
